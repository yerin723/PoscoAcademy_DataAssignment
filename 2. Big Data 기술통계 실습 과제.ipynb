# 기술통계 실습 2 (p. 217)
- 자동차의 모델과 연비(mpg)에 대한 자료를 조사한 Table이다.
- 이 범주형 변수인 구동방식, 모델에 대해 counts, percents, cumulative counts, cumulative percents를 계산하시오.

```python
df_mycars = pd.read_csv('/home/piai/다운로드/실습화일/1 Big Data 분석 기초/mycars.csv') #데이터 가져오기
```

```python
df = df_mycars[['driving', 'model']] #1. 모델 Table
df
```

```python
count = df_mycars['model'].value_counts().sort_index()
count.head() #차례대로 5개만 출력
```

```python
cumcnt = np.cumsum(count)
cumcnt.head()
```

```python
percent = count/sum(count)*100
percent.head()
```

```python
cumpct = np.cumsum(percent)
cumpct.head()
```

```python
count_data = pd.DataFrame({'Count': count, 'CumCnt':cumcnt,'Percent':percent, 'CumPct':cumpct})
count_data.columns.name='model'
count_data
```

```python
df = df_mycars[['driving', 'mpg']] #2. 연비 Table
df
```

```python
count = df['driving'].value_counts().sort_index()
count
```

```python
cumcnt = np.cumsum(count)
cumcnt
```

```python
percent = count/sum(count)*100
percent
```

# 기술통계 실습 3 (p. 230)
- 대한민국 남성의 20~30대 키를 측정한 결과, 평균 175cm, 표준편차 5cm이다.
- 나의 키가 180cm일 경우에 나는 몇 % 범위에 위치할까요?

```python
#누적확률 값 계산
mu = 175
sigma = 5
x = 180
prob = stats.norm.cdf(x, mu, sigma) # 정규 분포의 누적 확률 값 계산
print('나의 키가 {}cm 일 경우에 나는 {:.2f}% 범위에 위치한다'.format(x, (1-prob)*100))
```

# 기술통계 실습 4 (p. 233)
- 어떤 자동차 1리터당 주행거리가 평균 12km, 표준편차 3km인 정규분포를 이룬다.
1) 1리터를 가지고 9km 이상 15km 이하를 달릴 확률은?
2) 1리터를 가지고 15km 이상을 달릴 확률은?

```python
mu = 12
sigma = 3
x = 15
prob = stats.norm.cdf(x, mu, sigma)
print('{}km 이상 {}km 이하를 달릴 확률은 {:.2f}이다.'.format(9, 15, (prob-0.5)*2))
print('{}km 이상 달릴 확률은 {:.2f}이다.'.format(15, (1-prob)))
```

# 기술통계 실습 5 (p. 240)
- 확률변수 t는 자유도가 100인 t-분포에 따른다.
- t값이 2.0일 때 P(T <= t)를 계산하시오.

```python
df = 100
t = 2
prob = stats.t.cdf(t,df)
print('P(T <= F): {:.3f}'.format(prob))
```

# 기술통계 실습 6 (p. 243)
- 확률변수 x^2는 자유도가 8인 x^2 분포를 따른다.
- x^2값이 3.0일 때 P(X <= x^2)를 계산하시오.

```python
chisq = 3.0
df = 8
prob = stats.chi2.cdf(chisq,df)
print('P(X <= {}):{:.4f}'.format(chisq,prob))
```

# 기술통계 실습 7 (p. 247)
- 확률변수 F는 각각 자유도가 10, 100인 F분포를 따른다.
- F값이 1.5일 때, P(X <= F)를 계산하시오.

```python
f = 1.5
dfnum = 10
dfden = 100
prob = stats.f.cdf(f,dfnum,dfden)
print('P(X <= F): {:.3f}'.format(prob))
```
