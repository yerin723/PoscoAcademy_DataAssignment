# 4. Python을 이용한 데이터 선택 실습 (p.36)
- 정규분포에서 추출한 10개의 무작위 데이터로 slice, loc, iloc, query 등을 이용하여 데이터를 선택하라
- 데이터: 정규분포에서 추출한 10개의 무작위 데이터 (변수: 4개, 자료 수: 10개, random seed = 1234)

```python
import pandas as pd #pandas : 데이터 프레임 연산용
import numpy as np #numpy : 숫자 연산용
from numpy.random import randn #random 함수

#step 1. 데이터 구성하기
np.random.seed(1234)
df = pd.DataFrame(randn(10,4), columns=['A','B','C','D']) #랜덤 함수를 이용해 df데이터 생성
df
```

```python
#step 2. slice
df['B'] #1) B열 선택: Series 반환
```

```python
df[['B','D']] #2) B, D열 선택: DataFrame 반환
```

```python
df[4:9] #3) 4~8행 선택
```

```python
#step 3-1. loc
df.loc[2:8, ['A','C']] #1) 2~8행, A,C열 선택
```

```python
df.loc[:,['A','C']] #2) A,C열 선택
```

```python
df.loc[3:8] #3) 3~8행 선택
```

```python
df.loc[3] #4) 3행 선택(단일행)
```

```python
#step 3-2. iloc
df.iloc[0:7, [0]] #1) 0~6행, A열 선택
```

```python
df.iloc[:, [0] ] #2) A열 선택
```

```python
df.iloc[2:7, :] #3) 2~6행 선택
```

```python
df.iloc[5:6, :] #4) 5행 선택(단일행)
```

```python
#step 4. query
df.query('D > 0.5') #1) D>0.5 데이터 선택
```

```python
df[df.B > df.D] #2) B>D 데이터 선택 => slice표현으로 선택
```

```python
df.query('A > 0.5 & D < 0.0') #3) A>0.5 그리고 D<0.0 데이터 선택
```

```python
df.query('A > 0.5 | D < 0.0') #4) A>0.5 혹은 D<0.0 데이터 선택
```

# 5. Python을 이용한 데이터 결합 실습 (p.50)
- 아래와 같은 샘플 데이터 df4, df5, df6를 이용해 concat(좌우 결합)을 실행하라.
- 데이터 : 직접 생성한 아래와 같은 df4, df5, df6 데이터(변수 4개, 자료수 4개)

```python
import pandas as pd
#step 1. df4, df5, df6 데이터 생성
df4 = pd.DataFrame({'A' : ['A0','A1','A2','A3'],
                   'B' : ['B0','B1','B2','B3'],
                   'C' : ['C0','C1','C2','C3'],
                   'D' : ['D0','D1','D2','D3']},
                  index = [0,1,2,3])

df5 = pd.DataFrame({'E' : ['E4','E5','E6','E7'],
                   'F' : ['F4','F5','F6','F7'],
                   'G' : ['G4','G5','G6','G7'],
                   'H' : ['H4','H5','H6','H7']},
                  index = [0,1,2,3])
df6 = pd.DataFrame({'I' : ['I8','I9','I10','I11'],
                   'J' : ['J8','J9','J10','J11'],
                   'K' : ['K8','K9','K10','K11'],
                   'L' : ['L8','L9','L10','L11']},
                  index = [0,1,2,3])
                   
#step 2~3. concat을 이용한 데이터 결합 및 parameter 변경
result = pd.concat([df4,df5,df6], axis = 1) #좌우 결합 : axis = 1
result
```

# 7. Python을 이용한 데이터 변환 실습 (p.72)
- pivot을 이용하여 GENDER와 AGEGROUP별 WEIGHT의 평균을 아래와 같이 변환하라.
- 데이터 : FITNESS.CSV(변수: 10개, 자료수: 31개)

```python
import pandas as pd
import numpy as np
#Step 1. 혈당 산소 요구량 파일 읽어오기
df = pd.read_csv('/home/piai/다운로드/실습화일/1 Big Data 분석 기초/FITNESS.csv', encoding = 'euc-kr')
```

```python
#step 2 - 1) GENDER별 WEIGHT 평균 계산
df.groupby('GENDER')['WEIGHT'].agg('mean') #groupby, agg를 이용한 평균값 계산
```

```python
#step 2 - 2) AGEGROUP별 WEIGHT 평균 계산
df.groupby(['GENDER', 'AGEGROUP'])['WEIGHT'].agg('mean') #groupby, agg를 이용한 평균값 계산
```

```python
#step 3. Pivot을 이용한 변환
df.groupby(['GENDER', 'AGEGROUP'],as_index=False).agg('mean') #groupby, agg를 이용한 평균값 계산
```

```python
pivot = pd.pivot_table(df, values = 'WEIGHT', index = ['GENDER'], columns = ['AGEGROUP'])
pivot
```
